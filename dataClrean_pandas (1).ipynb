```python
def Gconvertor(x):
  if x=='Female' or  x=='female' or  x=='F':
    return 'f'
  elif x=='Male' or  x=='male' or  x=='MALE' or x=='M':
    return 'm'
  else :
    return x
```

```python
df['Gender']=df['Gender'].apply(Gconvertor)
```

```python
s1=df['Age'].quantile(0.25)
s2=df['Age'].quantile(0.75)
```

```python
# find outlier

df[(df['Age']<lower) | (df['Age'] > upper)]
```

```python
((df['Age']<lower) | (df['Age'] > upper)).sum()
```

```python
# remove outlier
df[(df['Age']>lower) & (df['Age'] < upper)]
```
