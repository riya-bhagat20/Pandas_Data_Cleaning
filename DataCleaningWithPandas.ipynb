```python
# I. Write a program to create a DataFrame named df from the
# above dictionary data, using labels as the index.


import pandas as pd
import numpy as np
data = {
'creatures': ['Eagles','Eagles','Ducks','Herons','Herons','Eagles','Ducks','Eagles','Herons','Herons','Eagles'],
'weight': [2.0,5,3.0,np.nan,7,2.5,6,np.nan,9,3,2.0],
'frequency': [1,3,2,5,2,3,1,3,4,2,1],
'status': ['no','no','yes',np.nan,'yes','yes','yes','no','yes','yes','no'],
'height': [1.2,2.5,1.8,3.0,2.2,1.9,2.6,1.4,2.8,2.3,1.1],
'region': ['North','South','East','West','North','South','East','West','North','East','South'],
'danger_level': ['low','medium','high','low','high','medium','high','low','medium','high','low']
}

labels = ['x','y','z','w','v','u','t','s','r','q','p']

df = pd.DataFrame(data, index=labels)
print(df)
```

```python
# II. Write a program to display the basic information of the
# DataFrame, including its summary statistics and data types.

df.info()
```

```python
# III. Write a program to display every alternate row of the
# DataFrame (starting from index 0).

df.iloc[::2]
```

```python

# IV. Write a program to display only rows at index positions 1st,
# 3rd, and 7th from columns creatures and weight of the
# DataFrame.
df.iloc[[1,3,7]][['creatures','weight']]
```

```python
# V. Write a program to select and display the rows where the
# frequency is less than 4.
result = df[df['frequency'] < 4]
print(result)

```

```python
# VI. Write a program to select and display all rows where there are
# NaN values in the columns weight or status.

result = df[df['weight'].isna() | df['status'].isna()]
print(result)
```

```python
# VII. Write a program to fill the NaN values in the DataFrame with
# the mode (most frequent value) of the respective columns.

df = df.fillna(df.mode().iloc[0])
print(df)
```

```python
# VIII. Write a program to find the total number of frequency visits for
# the bird 'Herons'.

total = df[df['creatures'] == 'Herons']['frequency'].sum()
print("Total frequency visits for Herons:", total)
```

```python

# IX.
# Write a program to count and display the number of each type
# of creature in the DataFrame.
count = df['creatures'].value_counts()
print(count)
```

```python
# X. Write a program to remove duplicate rows from the DataFrame
# df permanently. Display the DataFrame after the changes.

df.drop_duplicates(inplace=True)
print(df)
```

```python
#xi Write a program to replace the status of 'no' to 'not available'
# for all occurrences in the DataFrame df.

df['status'] = df['status'].replace('no', 'not available')
print(df)
```

```python
# Write a program to add a new column size to the DataFrame df,
# where the values are based on the height column: label any
# bird taller than 2.5 as 'Large', otherwise label as 'Small'.
df['size'] = df['height'].apply(lambda x: 'Large' if x > 2.5 else 'Small')
print(df)
```

```python
# Write a program to sort the DataFrame df by danger_level and
# height in descending order.

df_sorted = df.sort_values(by=['danger_level','height'], ascending=False)
print(df_sorted)
```

```python
# XIV. Write a program to reset the index of the DataFrame df to the
# default integer index, without affecting the original index.

df_reset = df.reset_index()
print(df_reset)

```

```python
# XV. Write a program to create a new DataFrame df_copy by
# selecting only the rows where weight is greater than 5, and
# include only the columns creatures, weight, and region.

df_copy = df[df['weight'] > 5][['creatures','weight','region']]
print(df_copy)
```
